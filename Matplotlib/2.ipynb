```python
import matplotlib.pyplot as plt
```

```python
x = ["p","j","c","c++"]
y = [85,70,60,55]
z = [20,30,40,50]
c = ["y","b","m","g"]
c1 = ["g","m","b","y"]
plt.bar(x,y,width = 0.5,color = c,edgecolor = "black",linewidth = 1.5,linestyle = ":")
plt.bar(x,z,width = 0.5,color = c1,edgecolor = "black",linewidth = 1.5,linestyle = ":")
plt.xlabel("language",fontsize = 14,color = "r")
plt.ylabel("popularity",fontsize = 14,color = "r")
plt.title("Language and Popularity") 
plt.show()
 
```

```python
import numpy as np

x = ["p","j","c","c++"]
y = [85,70,60,55]
z = [20,30,40,50]
c = ["y","b","m","g"]
c1 = ["g","m","b","y"]

width = 0.2
p = np.arange(len(x))
p1 = [j++width for j in p]

plt.barh(p,y,width,color = c,edgecolor = "black",linewidth = 1.5,linestyle = ":")
plt.barh(p1,z,width,color = c1,edgecolor = "black",linewidth = 1.5,linestyle = ":")

plt.xlabel("language",fontsize = 14,color = "r")
plt.ylabel("popularity",fontsize = 14,color = "r")
plt.title("Language and Popularity") 

plt.xticks(p+width/2,x)
plt.show()
 
```
