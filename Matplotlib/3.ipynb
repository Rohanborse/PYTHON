## Scatter

```python
import matplotlib.pyplot as plt
```

```python
x = [4,5,7,6,4,6,7,8.7,9,6,5,4,3,5,6,8,9,6,5,4]
y = [7,9,5,3,5,7,9,7,6,4,3,4,6,7,9,7,6,5,4,5]
z = [3,5,6,9,9,8,7,5,7,8,9,8,6,5,4,6,7,8,9,4]
c = ["r","b","g","m","r","b","g","m","r","b","g","m","r","b","g","m","r","b","g","m"]

size = [500,600,400,900,900,800,7000,500,700,8000,9000,800,600,500,400,600,700,800,9000,400]
plt.scatter(x,y,color = c,s = size,alpha = 0.3)
plt.scatter(x,z,color = c,s = size,alpha = 0.4)
plt.colorbar()

plt.show()
```

```python
x= [50,60,91,70,80]
y = [30,40,50,60,60]
z =[20,30,40,50,20]
plt.bar(x,y,color = "orange",width = 4)
plt.bar(x,z,color = "r",width = 8)

plt.show()
```

# Histrogram

```python
import numpy as np
import random
```

```python
y = np.random.randint(0,500,(400))
plt.hist(y,color = "#FF44CC",bins = 30,histtype = "step",orientation = "vertical",label="random") ## 
plt.axvline(245,color = "r",label = "line",alpha = 0.4)
plt.legend()
plt.grid()
plt.show()
```

# pia chart

```python
x =[10,20,30,50]
y =["j","p","c","c++"]
ex = [1.4,0.0,0.0,0.0]
plt.pie(x,labels = y,explode=ex,autopct="%0.1f%%",shadow = True,radius =1.5,startangle=30,
        textprops={"fontsize":15},counterclock=True,wedgeprops={"linewidth":9})
plt.show()   
```

```python
plt.pie([1])
plt.show()
```

```python
x = [10,20,30,40]
x1 = [40,30,20,10]

plt.pie(x)
plt.pie(x1,radius = 0.7,colors = "w")
```

# stem plt 

```python
x = [1,3,4,6,8,9]
y = [9,8,6,5,4,6]
plt.stem(x,y,linefmt= ":",markerfmt = "r+",basefmt = "g:",label = "python")
plt.legend()
plt.show()
```

# box plot

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70,80,90,200]
plt.boxplot(x,notch = True,vert = False,widths= 0.3,labels = ["python"],patch_artist = True,showmeans = True,whis = 2
            ,sym = "g+")
plt.show()
```
