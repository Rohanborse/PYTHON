## Dropna 

```python
import pandas as pd
```

```python
var = pd.read_csv("..\\data\\Data.csv")
```

```python
var
```

```python
var.dropna()
                      ## this will Drop the row that have NaN value
```

```python
var.dropna(axis=1)            ## this will Drop the coloumn that have NaN value   
                              ##   beacause we use axis = 1 it is working along the coloumn
```

```python
var.dropna(how="all")                 ## if total row have value NaN that row will only remove
var.dropna(how="any")                 ## if any NaN value is present then total row will be remove
```

```python
var.dropna(subset=["paperCount"])  ## this will search for the coloum paperCount and in that coloum 
                                   ## if there is NaN value is present then that coloumn will be Remove
```
