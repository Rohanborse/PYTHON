# Merge

```python
import pandas as pd
```

```python
var = pd.DataFrame( {"A":[1,2,3,4],"B":[5,6,7,8]})
var1 = pd.DataFrame( {"A":[1,2,3,4],"C":[9,10,11,12]})

pd.merge(var,var1)   ## this will merge two dataframe in one DataFrame
```

```python
var2 = pd.DataFrame( {"A":[1,2,3,4],"B":[5,6,7,8]})
var3 = pd.DataFrame( {"A":[1,2,3,5],"C":[9,10,11,12]})

pd.merge(var2,var3,how = "outer" , indicator = True)    ## in how we can use right ,left, outer
```

```python
var3 = pd.DataFrame( {"A":[1,2,3,4],"B":[5,6,7,8]})
var4 = pd.DataFrame( {"A":[1,2,3,5],"B":[9,10,11,12]})

pd.merge(var3,var4,left_index=True,right_index=True,suffixes=("name","")) 
```

# Concat

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,22,33,44])
pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame( {"A":[1,2,3,4],"B":[5,6,7,8]})
d2 = pd.DataFrame( {"A":[1,2,3,5],"B":[9,10,11,12]})
pd.concat([d1,d2],axis=0)
```

```python
d1 = pd.DataFrame( {"A":[1,2,3,4],"B":[5,6,7,8]})
d2 = pd.DataFrame( {"A":[1,2,3,5],"B":[9,10,11,12]})
pd.concat([d1,d2],axis=1)
```

```python
d3 = pd.DataFrame( {"A":[1,2,3,4],"B":[5,6,7,8]})
d4 = pd.DataFrame( {"A":[1,2],"B":[9,10]})
pd.concat([d3,d4],axis=1,join = "inner")
```

```python
d1 = pd.DataFrame( {"A":[1,2,3,4],"B":[5,6,7,8]})
d2 = pd.DataFrame( {"A":[1,2,3,5],"B":[9,10,11,12]})
pd.concat([d1,d2],axis=0,keys = ["d1","d2"])
```

```python
d1 = pd.DataFrame( {"C":[1,2,3,4]})
d2 = pd.DataFrame( {"A":[1,2,3,5],"B":[9,10,11,12]})
pd.concat([d1,d2],axis=0)
```
