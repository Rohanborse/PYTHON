# Replace

to remove the NAN means not a number we have to use "np.nan"

```python
import pandas as pd
import numpy as np
```

```python
csv = pd.read_csv("..\\data\\pldb.csv",low_memory = False)
csv
```

```python
csv.replace(to_replace=1,value= "0")        ## Replace the Value
```

```python
csv.replace("[A-Za-z]","0",regex = True)  ## replace the total data from a-z by zero
```

```python
csv.replace({"type":"[a-z]"},22,regex = True)   ## this will only change the value of the "type" coloumn into 22
```

```python
csv.replace(0,method = "ffill")      ## in this we can change the data of any string or 
                                       ##integer by its forword data or Backword Data
                                                ## For forword we use      method = "ffill"
                                                ## For backword we use     method = "bfill"
```

```python
csv.replace(0,method="ffill",inplace=True)
csv
```
