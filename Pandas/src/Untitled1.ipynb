```python
import pandas as pd
iris_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
iris_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data_dict = {'name': ['p1', 'p2', 'p3', 'p4', 'p5', 'p6'],
             'age': [20, 20, 21, 20, 21, 20],
             'math_marks': [100, 90, 91, 98, 92, 95],
             'physics_marks': [90, 100, 91, 92, 98, 95],
             'chem_marks': [93, 89, 99, 92, 94, 92]
             }
 

df = pd.DataFrame(data_dict)
 

df
```

```python
df.plot(kind='bar',
        x='math_marks',
        y='physics_marks',
        color='red')
 

plt.title('ScatterPlot')
 
# show the plot
plt.show()
```

```python
ax = plt.gca()
df.plot(kind='line', x='name',
        y=['chem_marks','math_marks','physics_marks'],
        color=['black','green','red'])
plt.title('LinePlots')
plt.show()
```
